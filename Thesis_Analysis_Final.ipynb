```python
import pandas as pd
import numpy as np

"""**Data Read**"""

commercialpresent=pd.read_csv('drive/MyDrive/Present/CommercialPresent.txt')

commercialmrt=pd.read_csv('drive/MyDrive/MRT/CommercialMRT.txt')

commercialproposed=pd.read_csv('drive/MyDrive/Proposed/CommercialProposed.txt')

"""**Maximum Travel Distance**"""

com_present_max=commercialpresent['Total_Time'].max()
com_present_max

com_mrt_max=commercialmrt['Total_Time'].max()
com_mrt_max

com_proposed_max=commercialproposed['Total_Time'].max()
com_proposed_max


"""**Insert New Columns**"""

commercialmrt.insert(5,"Time", True)
commercialmrt.insert(6,"BetaTime","Na")
commercialmrt.insert(7,"Accessibility","NA")

commercialpresent.insert(5,"Time", True)
commercialpresent.insert(6,"BetaTime","Na")
commercialpresent.insert(7,"Accessibility","NA")

commercialproposed.insert(5,"Time", True)
commercialproposed.insert(6,"BetaTime","Na")
commercialproposed.insert(7,"Accessibility","NA")


"""**Creating Lables**"""

labels = ["{0}-{1}".format(i, i + 4) for i in range(0, 240, 5)]

plabels = ["{0}-{1}".format(i, i + 4) for i in range(0, 360, 5)]

"""**Categorized the Travel Time Data**"""

commercialpresent['Time']=pd.cut(commercialpresent.Total_Time,range(0,365,5),right=False,labels=plabels)
commercialpresent

commercialmrt["Time"] = pd.cut(commercialmrt.Total_Time, range(0, 245, 5), right=False, labels=labels)
commercialmrt

commercialproposed["Time"] = pd.cut(commercialproposed.Total_Time, range(0, 245, 5), right=False, labels=labels)
commercialproposed

"""**Exporting the Travel Time Data**"""

com_p_time=commercialpresent['Time'].value_counts()
com_p_time.to_csv(r"drive/MyDrive/Data/com_p_time.csv")


com_m_time=commercialmrt['Time'].value_counts()
com_m_time.to_csv(r"drive/MyDrive/Data/com_m_time.csv")

com_pr_time=commercialproposed['Time'].value_counts()
com_pr_time.to_csv(r"drive/MyDrive/Data/com_pr_time.csv")


"""**Beta Value Calculation**"""


com_p_beta= -np.log(0.5)/commercialpresent['Total_Time'].median()
com_p_beta

com_m_beta= -np.log(0.5)/commercialmrt['Total_Time'].median()
com_m_beta

com_pr_beta = -np.log(0.5)/commercialproposed['Total_Time'].median()
com_pr_beta


"""**Beta Multiply Travel Time**"""


commercialpresent['BetaTime']= commercialpresent['Total_Time']*com_p_beta
commercialpresent


commercialmrt['BetaTime']= commercialmrt['Total_Time']*com_m_beta
commercialmrt


commercialproposed['BetaTime']= commercialproposed['Total_Time']*com_pr_beta
commercialproposed


"""**Accessibility Calculation**"""

commercialpresent['Accessibility']= commercialpresent['FloorArea']*np.exp(-commercialpresent['BetaTime'])
commercialpresent

commercialmrt['Accessibility']= commercialmrt['FloorArea']*np.exp(-commercialmrt['BetaTime'])
commercialmrt

commercialproposed['Accessibility']= commercialproposed['FloorArea']*np.exp(-commercialproposed['BetaTime'])
commercialproposed


"""**Accessibility Grouped by OriginID**"""

com_p_access=commercialpresent['Accessibility'].groupby(commercialpresent['OriginID']).sum()
com_p_access.to_csv(r"drive/MyDrive/Data/com_p_access.csv")

com_m_access=commercialmrt['Accessibility'].groupby(commercialmrt['OriginID']).sum()
com_m_access.to_csv(r"drive/MyDrive/Data/com_m_access.csv")

com_pr_access=commercialproposed['Accessibility'].groupby(commercialproposed['OriginID']).sum()
com_pr_access.to_csv(r"drive/MyDrive/Data/com_pr_access.csv")
```
